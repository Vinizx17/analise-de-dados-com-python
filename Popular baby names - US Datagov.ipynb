```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Popular_Baby_Names.csv')

display(df)
```

```python
top_names = df.groupby('Child\'s First Name')['Count'].sum().sort_values(ascending=False).head(15)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_names.values, y=top_names.index, palette='viridis')
plt.title('Top 15 Nomes Mais Populares')
plt.xlabel('Contagem Total')
plt.ylabel('Nome')
plt.show()
```
