```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('LakeCounty_Health_-6821310239639457742.csv')

display(df)
```

```python
# Definir as regies dos estados (por exemplo, dividir os estados em quatro regies)
regioes = {
    'Nordeste': ['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'New Jersey', 'Pennsylvania'],
    'Sudeste': ['Delaware', 'Maryland', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida'],
    'Meio-Oeste': ['Ohio', 'Indiana', 'Illinois', 'Michigan', 'Wisconsin', 'Minnesota', 'Iowa', 'Missouri', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas'],
    'Oeste': ['Montana', 'Idaho', 'Wyoming', 'Colorado', 'New Mexico', 'Arizona', 'Utah', 'Nevada', 'California', 'Oregon', 'Washington', 'Alaska', 'Hawaii']
}

# Mapear os estados para as regies
def mapear_regiao(estado):
    for regiao, estados_regiao in regioes.items():
        if estado in estados_regiao:
            return regiao
    return 'Puerto Rico'

# Adicionar uma coluna de regio ao DataFrame
df['Regio'] = df['NAME'].apply(mapear_regiao)

# Contar o nmero de estados em cada regio
contagem_regiao = df['Regio'].value_counts()

# Criar o grfico de pizza
plt.figure(figsize=(8, 6))
contagem_regiao.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral', 'lightsalmon', 'lightpink'])
plt.title('Distribuio de Estados por Regio')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
